```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
%matplotlib inline
```

```python
data = pd.read_csv('white_house_tweet_data.csv')
```

```python
data['created_at'] = pd.to_datetime(data.created_at)
```

```python
data.head()
```

```python
data['created_at_est'] = pd.DatetimeIndex(data['created_at']).tz_localize('UTC').tz_convert('US/Eastern')
```

```python
data[data.created_at >= '2018-07-13'][['created_at']].groupby(data["created_at"].dt.date).count().plot(kind="line")
```

```python
data[data.created_at_est >= '2018-07-13'][['created_at_est']].groupby(data["created_at_est"].dt.hour).count().plot(kind="bar")
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in list(set(data['created_at_est'].dt.weekday_name)):
#     new_plot = 
#     ax = fig.add_subplot()
#     ax.text(0.5, 0.5, i,
#            fontsize=18, ha='center')
    data[(data.created_at_est >= '2018-07-13') & (data.created_at_est.dt.weekday_name == i)][['created_at_est']].groupby(data["created_at_est"].dt.hour).count().plot(kind="bar", title = i)
```
